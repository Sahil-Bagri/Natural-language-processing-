```python
import nbformat

def get_notebook_content(notebook_path):
    with open(notebook_path, 'r', encoding='utf-8') as f:
        nb = nbformat.read(f, as_version=4)
    
    content = []
    for cell in nb.cells:
        if cell.cell_type == 'code':
            content.append(cell.source)
        elif cell.cell_type == 'markdown':
            content.append(f"'''{cell.source}'''")
    
    return '\n\n'.join(content)

notebook_path = r"C:\Users\Sahil\Nlp suneet\06-01-compare-both.ipynb"  # replace with your notebook path
print(get_notebook_content(notebook_path))
```

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
words = ['charging' , 'charges', 'charged', 'study', 'studying', 'easliy', 'fairly','losses', 'was' , 'neccessary ', 'implementation ', 'permission' , 'learn']
```

```python
for word in words:
  print(word + ' : ' + ps.stem(word))
```

```python
sentence = "police is charging the support a sit ran away".split()
```

```python
sentence
```

```python
for word in sentence:
  print(word + ' : ' + ps.stem(word))
```

```python
from nltk.stem.snowball import SnowballStemmer
snowball = SnowballStemmer(language='english')
```

```python
for word in words:
    print(f"{word} : {snowball.stem(word)} ")
```
