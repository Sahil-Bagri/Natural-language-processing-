```python
import nbformat

def get_notebook_content(notebook_path):
    with open(notebook_path, 'r', encoding='utf-8') as f:
        nb = nbformat.read(f, as_version=4)
    
    content = []
    for cell in nb.cells:
        if cell.cell_type == 'code':
            content.append(cell.source)
        elif cell.cell_type == 'markdown':
            content.append(f"'''{cell.source}'''")
    
    return '\n\n'.join(content)

notebook_path = r"C:\Users\Sahil\Nlp suneet\08-vocabulary & matching.ipynb"  # replace with your notebook path
print(get_notebook_content(notebook_path))
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
from spacy.matcher import Matcher
matcher =Matcher(nlp.vocab)
```

```python
pattern1 = [{'LOWER':'solarpower'}]
pattern2 = [{'LOWER':'Solar'},{'LOWER':'Power'}]
pattern3 = [{'LOWER':'Solar'},{'IS_PUNCT':True},{'LOWER':'Power'}]

matcher.add('SolorPower' , [pattern1 , pattern2, pattern3])
```

```python
doc = nlp(u'The Solar Power industry continues to grow as demand \ for solarpower increases. Solar-Power cars are gaining popularity')
```

```python
found_matches = matcher(doc)
```

```python
found_matches
```

```python
for match_id , start , end in found_matches:
  string_id = nlp.vocab.strings[match_id]
  span = doc [start:end]
  print(match_id , string_id , start , end , span.text)
```

```python
pattern1 = [{'LOWER':'solarpower'}]
pattern2 = [{'LOWER':'Solar'},{'IS_PUNCT':True , 'OP':'*'},{'LOWER':'Power'}]

matcher.add('solorPower' , [pattern1 , pattern2])
```

```python
from spacy.matcher import PhraseMatcher
matcher = PhraseMatcher(nlp.vocab)
```
